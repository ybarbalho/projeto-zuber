### 3. Estudando casos hipotéticos

```python
# Importandos os dados climáticos
df_weather = pd.read_csv('./datasets/project_sql_result_07.csv')
```

Acima obtemos uma tabela que contém informações climáticas, data e horário de coleta e duração das viagens do período estudado.

A partir disso, iremos trabalhar com a análise das viagens em períodos chuvosos. Sendo assim, filtraremos nossa tabela para obter os dados dos dias cuja condição climática foi definida como ruim em nosso registro de viagens (valor definido como *Bad*).

```python
# Filtrando a nossa tabela
df_rain = df_weather[df_weather['weather_conditions']=='Bad']
```

```python
# Observando os dados
df_rain.head()
```

### 4. Testando a hipótese: sábados chuvosos

Agora vamos testar a hipótese aplicada de que a duração média dos passeios do Loop para o Aeroporto Internacional O'Hare muda durante os sábados chuvosos. Por se tratar da afirmação em caso, essa será a hipótese nula do nosso estudo.

Dessa forma, nossa hipótese alternativa será de que a duração média entre o Loop e o Aeroporto Internacional O'Hare não sofre alteração durante os sábados de chuva.

```python
# Calculando a variância de duração das corridas em sábados chuvosos.
df_rain['duration_seconds'].var()
```

```python
# Calculando a variância de duração média das corridas de sábado.
df_weather['duration_seconds'].var()
```

Calculamos as variâncias das variáveis em estudo para descobrir qual teste deverá ser aplicado à hipótese em estudo.

No caso acima, notamos proximidade nos valores das variâncias do teste de hipótese a ser aplicado. Portanto, o recomendável é que utilizemos o teste t de Student na análise da nossa hipótese nula, e atribuíremos o valor padrão de 5% (0.05) como nível de significância.

```python
# Testando a hipótese: duração média dos passeios do Loop para o Aeroporto Internacional O'Hare muda durante os sábados chuvosos.

alpha = .05 # Atribuímos 0.05 como valor do nível de significância.
results = st.ttest_ind(df_rain['duration_seconds'], df_weather['duration_seconds']) # Aplicamos as médias em estudo, considerando a proximidade das variâncias.

print(f'valor-p: {results.pvalue}')

if (results.pvalue >= alpha): # Comparamos o valor-p com o nível de significância para avaliar a hipótese
    print("Rejeitamos a hipótese nula")
else:
    print("Não podemos rejeitar a hipótese nula")
```

Analisando o valor-p, observamos que o valor está bem abaixo do nosso nível de significância (de 0.05), o que indica que não podemos rejeitar a nossa hipótese nula.

Rejeitando a hipótese alternativa, identificamos que possivelmente existe alteração na duração dos passeios do Loop para o Aeroporto Internacional O'Hare durante sábados chuvosos, o que era de se esperar, visto que dias chuvosos causam dificuldades no fluxo do tráfego.
